```python


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create Student Data
data = {
    'Name': ['Arun', 'Priya', 'Karthik', 'Divya', 'Rahul'],
    'Maths': [85, 45, 92, 30, 70],
    'Science': [78, 55, 89, 40, 75],
    'English': [90, 60, 95, 35, 80],
    'Computer': [88, 50, 96, 45, 72]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Total Marks
df['Total'] = df[['Maths', 'Science', 'English', 'Computer']].sum(axis=1)

# Calculate Average Marks
df['Average'] = df['Total'] / 4

# Grade Assignment Function
def grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 75:
        return 'B'
    elif avg >= 50:
        return 'C'
    else:
        return 'D'

# Assign Grades
df['Grade'] = df['Average'].apply(grade)

# Pass/Fail Status
df['Result'] = df['Average'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')

# Display Full Data
print("Student Performance Table:")
print(df)

# Identify Topper
topper = df[df['Average'] == df['Average'].max()]
print("\nTopper:")
print(topper[['Name', 'Average']])

# Identify Weak Students
weak_students = df[df['Average'] < 50]
print("\nWeak Students:")
print(weak_students[['Name', 'Average']])

# Subject-wise Average
subject_avg = df[['Maths', 'Science', 'English', 'Computer']].mean()
print("\nSubject-wise Average:")
print(subject_avg)

# Highest Marks in Each Subject
highest_marks = df[['Maths', 'Science', 'English', 'Computer']].max()
print("\nHighest Marks in Each Subject:")
print(highest_marks)

# Lowest Marks in Each Subject
lowest_marks = df[['Maths', 'Science', 'English', 'Computer']].min()
print("\nLowest Marks in Each Subject:")
print(lowest_marks)

# Bar Chart for Student Average
df.plot(x='Name', y='Average', kind='bar')
plt.title("Student Average Marks")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()

# Pie Chart for Pass/Fail
df['Result'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pass vs Fail")
plt.ylabel("")
plt.show()

# Subject-wise Performance Chart
subject_avg.plot(kind='bar')
plt.title("Subject Average Performance")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()

# Line Chart for Student Subject Performance
df[['Maths', 'Science', 'English', 'Computer']].T.plot(kind='line')
plt.title("Student Subject Performance")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
